```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the weather data
df = pd.read_csv('weather_data.csv')

# Visualize the average temperature in each location
sns.lineplot(x='date', y='temperature', hue='location', data=df)
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Average Temperature by Location')
plt.show()

# Visualize the distribution of precipitation in each location
sns.histplot(x='precipitation', hue='location', data=df)
plt.xlabel('Precipitation (mm)')
plt.ylabel('Frequency')
plt.title('Distribution of Precipitation by Location')
plt.show()

# Create a correlation matrix to visualize the correlation between weather variables
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix of Weather Variables')
plt.show()
```
